```python
import pandas as pd

# -----------------------------
# Step 1: Define get_periods()
# -----------------------------
def get_periods(freq, analysis_date):
    if freq == 'Monthly':
        months = [analysis_date - pd.DateOffset(months=i) for i in range(3)][::-1]  # oldest to newest
        return months
    
    elif freq == 'Quarterly':
        curr_start = (analysis_date - pd.DateOffset(months=2)).replace(day=1)
        curr_end = analysis_date
        prev_start = (curr_start - pd.DateOffset(months=3)).replace(day=1)
        prev_end = curr_start - pd.Timedelta(days=1)
        return [(curr_start, curr_end), (prev_start, prev_end)]
    
    elif freq == 'Half-Yearly':
        curr_start = (analysis_date - pd.DateOffset(months=5)).replace(day=1)
        curr_end = analysis_date
        prev_start = (curr_start - pd.DateOffset(months=6)).replace(day=1)
        prev_end = curr_start - pd.Timedelta(days=1)
        return [(curr_start, curr_end), (prev_start, prev_end)]
    
    elif freq == 'Yearly':
        curr_start = (analysis_date - pd.DateOffset(months=11)).replace(day=1)
        curr_end = analysis_date
        prev_start = (curr_start - pd.DateOffset(months=12)).replace(day=1)
        prev_end = curr_start - pd.Timedelta(days=1)
        return [(curr_start, curr_end), (prev_start, prev_end)]

# -----------------------------
# Step 2: Churn Calculation Function
# -----------------------------
def calculate_churn(customer_id, freq):
    if freq == 'Monthly':
        threshold = 0.25
        months = get_periods(freq, latest_date)
        monthly_sales = []
        
        for month_date in months:
            month_start = month_date.replace(day=1)
            month_end = (month_start + pd.offsets.MonthEnd(0))
            
            sales = new_data1[
                (new_data1['Customer'] == customer_id) &
                (new_data1['Fiscal_Period'] >= month_start) &
                (new_data1['Fiscal_Period'] <= month_end)
            ]["Sales_Volume(KG)'"].sum()
            
            monthly_sales.append(sales)
        
        if monthly_sales[-1] == 0:
            churn = 1
        else:
            drop1 = monthly_sales[1] < monthly_sales[0] * (1 - threshold)
            drop2 = monthly_sales[2] < monthly_sales[1] * (1 - threshold)
            churn = 1 if drop1 and drop2 else 0
        
        return pd.Series([monthly_sales[-1], sum(monthly_sales[:-1]), churn])
    
    else:
        threshold = 0.40
        (curr_start, curr_end), (prev_start, prev_end) = get_periods(freq, latest_date)
        
        curr_sales = new_data1[
            (new_data1['Customer'] == customer_id) &
            (new_data1['Fiscal_Period'] >= curr_start) &
            (new_data1['Fiscal_Period'] <= curr_end)
        ]["Sales_Volume(KG)'"].sum()
        
        prev_sales = new_data1[
            (new_data1['Customer'] == customer_id) &
            (new_data1['Fiscal_Period'] >= prev_start) &
            (new_data1['Fiscal_Period'] <= prev_end)
        ]["Sales_Volume(KG)'"].sum()
        
        if curr_sales == 0:
            churn = 1
        elif prev_sales > 0 and (curr_sales / prev_sales) < (1 - threshold):
            churn = 1
        else:
            churn = 0
        
        return pd.Series([curr_sales, prev_sales, churn])

# -----------------------------
# Step 3: Apply Churn Calculation
# -----------------------------
latest_date = new_data1['Fiscal_Period'].max()

final_df[['Current_Sales', 'Previous_Sales', 'is_churn']] = final_df.apply(
    lambda row: calculate_churn(row['Customer'], row['customer_purchase_behavior']), axis=1
)

# -----------------------------
# Step 4: Final Churn Dataset
# -----------------------------
churn_df = final_df[['Customer', 'customer_purchase_behavior', 'Current_Sales', 'Previous_Sales', 'is_churn']].copy()
churn_df.rename(columns={'customer_purchase_behavior': 'Customer_Frequency'}, inplace=True)
churn_df = churn_df.reset_index(drop=True)

print("✅ Churn dataset shape:", churn_df.shape)
print(churn_df.head())
```

```python
# calculate the average revenue based on customer purchase behavior and store it in monthly average revenue, quarterly average revenue, half yearly average revenue and yearly average revenue columns in final_df
average_monthly_revenue_last12m=final_df.groupby('customer_purchase_behavior')['Monetary_value'].mean().get('Monthly', 0)
average_quarterly_revenue_last12m=final_df.groupby('customer_purchase_behavior')['Monetary_value'].mean().get('Quarterly', 0)
average_halfyearly_revenue_last12m=final_df.groupby('customer_purchase_behavior')['Monetary_value'].mean().get('Half-Yearly', 0)
average_yearly_revenue_last12m=final_df.groupby('customer_purchase_behavior')['Monetary_value'].mean().get('Yearly', 0)
```

```python
# now calculate the difference between each customer's revenue in last 12 months and average revenue based on their purchase behavior keep them as high,low,medium in a new column as revenue_generated
def revenue_generated(row):
    revenue = row['Monetary_value']
    behavior = row['customer_purchase_behavior']
    
    if behavior == 'Monthly':
        avg_revenue = average_monthly_revenue_last12m
    elif behavior == 'Quarterly':
        avg_revenue = average_quarterly_revenue_last12m
    elif behavior == 'Half-Yearly':
        avg_revenue = average_halfyearly_revenue_last12m
    elif behavior == 'Yearly':
        avg_revenue = average_yearly_revenue_last12m
    else:
        return 'Medium'  # Default case
    
    if revenue > avg_revenue * 1.1:
        return 'High'
    elif revenue < avg_revenue * 0.9:
        return 'Low'
    else:
        return 'Medium'

final_df['revenue_generated'] = final_df.apply(revenue_generated, axis=1)
```

```python
# do chi-square test for independence between churn and each categorical variable in categorical_cols_final and give the cols which are dependent on churn with p-value<0.05
from scipy.stats import chi2_contingency
dependent_cols = []
categorical_p_values = {}
for col in categorical_cols:
    contingency_table = pd.crosstab(final_df[col], final_df['churn'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    categorical_p_values[col] = p
    if p < 0.05:
        dependent_cols.append(col)
        print(f"Column '{col}' is dependent on churn (p-value: {p})")
```

```python
from hyperopt import hp, tpe, Trials, fmin, STATUS_OK
from sklearn.metrics import f1_score, precision_score, recall_score
from xgboost import XGBClassifier
import numpy as np

# --- Compute class imbalance ratio ---
neg, pos = np.bincount(y_train)
scale_pos_weight = neg / pos
print(f"scale_pos_weight = {scale_pos_weight:.2f}")

# --- Define search space ---
space = {
    'n_estimators': hp.choice('n_estimators', [200, 300, 400, 500]),
    'max_depth': hp.choice('max_depth', [4, 6, 8, 10]),
    'learning_rate': hp.uniform('learning_rate', 0.01, 0.2),
    'subsample': hp.uniform('subsample', 0.6, 1.0),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.6, 1.0),
    'gamma': hp.uniform('gamma', 0, 0.5),
    'min_child_weight': hp.choice('min_child_weight', [1, 3, 5]),
    'reg_lambda': hp.uniform('reg_lambda', 1.0, 3.0),
    'reg_alpha': hp.uniform('reg_alpha', 0, 1.0),
    'scale_pos_weight': hp.uniform('scale_pos_weight', scale_pos_weight*0.5, scale_pos_weight*1.5)
}

# --- Objective function ---
def objective(params):
    model = XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42,
        enable_categorical=True,
        **params
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Combine precision & recall for optimization
    precision = precision_score(y_test, y_pred, average='binary')
    recall = recall_score(y_test, y_pred, average='binary')
    f1 = f1_score(y_test, y_pred, average='binary')
    
    # Balanced metric: minimize (1 - F1)
    score = (precision + recall + f1) / 3  # combine equally
    return {'loss': 1 - score, 'status': STATUS_OK}

# --- Run optimization ---
trials = Trials()
best_params = fmin(
    fn=objective,
    space=space,
    algo=tpe.suggest,
    max_evals=50,
    trials=trials,
    rstate=np.random.default_rng(42)
)

print("\nBest Hyperparameters found:")
print(best_params)

# --- Rebuild model with best params ---
best_model = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42,
    enable_categorical=True,
    n_estimators=[200, 300, 400, 500][best_params['n_estimators']],
    max_depth=[4, 6, 8, 10][best_params['max_depth']],
    learning_rate=best_params['learning_rate'],
    subsample=best_params['subsample'],
    colsample_bytree=best_params['colsample_bytree'],
    gamma=best_params['gamma'],
    min_child_weight=[1, 3, 5][best_params['min_child_weight']],
    reg_lambda=best_params['reg_lambda'],
    reg_alpha=best_params['reg_alpha'],
    scale_pos_weight=best_params['scale_pos_weight']
)

best_model.fit(X_train, y_train)

# --- Evaluation ---
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

print("\nClassification Report (Train):")
print(classification_report(y_train, y_train_pred))
print("\nClassification Report (Test):")
print(classification_report(y_test, y_test_pred))

print("\nConfusion Matrix (Train):")
print(confusion_matrix(y_train, y_train_pred))
print("\nConfusion Matrix (Test):")
print(confusion_matrix(y_test, y_test_pred))
```
